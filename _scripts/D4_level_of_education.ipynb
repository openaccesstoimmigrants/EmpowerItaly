```python
import json
import os
```

```python
given_json = ''' {"Education Level":{"5":"ED0-2","6":"ED0-2","7":"ED0-2","8":"ED0-2","9":"ED0-2","25":"ED3_4","26":"ED3_4","27":"ED3_4","28":"ED3_4","29":"ED3_4","45":"ED5-8","46":"ED5-8","47":"ED5-8","48":"ED5-8","49":"ED5-8"},"Year":{"5":2018,"6":2019,"7":2020,"8":2021,"9":2022,"25":2018,"26":2019,"27":2020,"28":2021,"29":2022,"45":2018,"46":2019,"47":2020,"48":2021,"49":2022},"Value (%)":{"5":50.3,"6":50.1,"7":50.3,"8":49.2,"9":47.6,"25":37.0,"26":37.3,"27":37.5,"28":38.9,"29":40.2,"45":12.7,"46":12.6,"47":12.2,"48":11.9,"49":12.2}} '''

# Mapping of education level codes to labels
level_mapping = {
    'ED0-2': 'Primary School',
    'ED3_4': 'Secondary School',
    'ED5-8': 'Tertiary School'
}

# Parse the given JSON
parsed_json = json.loads(given_json)

# Initialize an empty list to store the transformed data
result = []

# Extract data from the parsed JSON, transform it, and map education levels
for key, value in parsed_json['Year'].items():
    result.append({
        'Year': value,
        'EducationLevel': level_mapping[parsed_json['Education Level'][key]],
        'Value': parsed_json['Value (%)'][key]
    })

# Convert the transformed data to JSON format
transformed_json = json.dumps(result, indent=2)
print(transformed_json)


# Save the new file

# Specify the folder path to save the JSON file
folder_path = "../_datasets/Clean/D4"

# Create the folder if it doesn't exist
os.makedirs(folder_path, exist_ok=True)

# Define the filename for the JSON file
filename = "level_of_education.json"

# Generate the file path
file_path = os.path.join(folder_path, filename)

# Save the resident Italian population JSON to the file
with open(file_path, "w") as file:
    file.write(transformed_json)

print(f"JSON data saved to: {file_path}")
```
