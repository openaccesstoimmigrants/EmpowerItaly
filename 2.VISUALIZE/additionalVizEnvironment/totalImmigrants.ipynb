# foreign population 2019-2023


```python
myDf1 = pd.read_csv('https://raw.githubusercontent.com/openaccesstoimmigrants/openaccesstoimmigrants/main/2.VISUALIZE/additionalVizEnvironment/data19-23.csv')
```

```python
myDf1
```

```python
myDf1.dtypes
```

```python
myDf1.drop(myDf1[myDf1['Age'] == 'TOTAL'].index, inplace=True)
myDf1.drop(myDf1[myDf1['Age'] == 'Y_GE100'].index, inplace=True)
```

```python
myDf1['Age'] = myDf1['Age'].astype(int)
```

```python
myDf1.dtypes
myDf1
```

```python
minor_value = 14
major_value = 65

# Filter data based on conditions
new = myDf1[(myDf1['Age'] > minor_value) & (myDf1['Age'] <= major_value)]
new
```

```python
male = new[new['Sex'] == 'Male']
# male.to_csv('male.csv', encoding = 'utf-8-sig')
# output # files.download('male.csv')
```

```python
female = new[new['Sex'] == 'Female']
# female.to_csv('female.csv', encoding = 'utf-8-sig')
# output # files.download('female.csv')
```

```python
# sums male
male19 = male[male['Year'] == 2019]
sum_male19 = male19['Quantity'].sum()

male20 = male[male['Year'] == 2020]
sum_male20 = male20['Quantity'].sum()

male21 = male[male['Year'] == 2021]
sum_male21 = male21['Quantity'].sum()

male22 = male[male['Year'] == 2022]
sum_male22 = male22['Quantity'].sum()
```

```python
# sums female
female19 = female[female['Year'] == 2019]
sum_female19 = female19['Quantity'].sum()

female20 = female[female['Year'] == 2020]
sum_female20 = female20['Quantity'].sum()

female21 = female[female['Year'] == 2021]
sum_female21 = female21['Quantity'].sum()

female22 = female[female['Year'] == 2022]
sum_female22 = female22['Quantity'].sum()
```

```python
# sums 2019
tot19 = sum_female19 + sum_male19
tot19
```

```python
# sums 2020
tot20 = sum_female20 + sum_male20
tot20
```

```python
# sums 2021
tot21 = sum_female21 + sum_male21
tot21
```

```python
# sums 2022
tot22 = sum_female22 + sum_male22
tot22
```

```python
columns = ['2019', '2020', '2021', '2022']
lines = [tot19, tot20, tot21, tot22]

df = pd.DataFrame(list(zip(columns, lines)), columns=['Year', 'Total'])
df
```

```python
df.to_csv('totalImmigrants.csv', index=False)
files.download('totalImmigrants.csv')
```
