```python
grouped_data = d0_total_filtered.groupby(['geo', 'TIME_PERIOD'])['OBS_VALUE'].sum().reset_index()

pivoted_data = grouped_data.pivot(index='geo', columns='TIME_PERIOD', values='OBS_VALUE')

pivoted_data.plot(kind='bar', stacked=True)
```

```python
# Filter the data for the desired year
desired_year = 2020  # Replace with your desired year
filtered_data = d0_total_filtered[d0_total_filtered['TIME_PERIOD'] == desired_year]

# Sort the data by the number of immigrants in descending order
sorted_data = filtered_data.sort_values('OBS_VALUE', ascending=True)

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 5)

# Add x, y gridlines
ax.grid(visible = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)

# Create the bar chart
bar_height = 0.6
bar_padding = 0.6
plt.barh(range(len(sorted_data)), sorted_data['OBS_VALUE'], height=bar_height, edgecolor='white')
plt.yticks(range(len(sorted_data)), sorted_data['geo'], fontsize=10)

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.1,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

# Customize the plot
plt.xlabel('Number of Immigrants')
plt.ylabel('Countries')
plt.title(f'Immigrants by Country in {desired_year}')

# Adjust spacing of elements on the y-axis
plt.subplots_adjust(left=0.2)

# Adjust spacing between the bars
plt.gca().set_ylim([-0.5 - bar_padding/2, len(sorted_data) - 0.5 + bar_padding/2])

# Display the plot
plt.show()
```

```python

# total number of immigrants in Italy from 2012 to 2021
d0_total_filtered_IT = d0_total_filtered[(d0_total_filtered['geo'] == 'IT')]
immigrants_per_year_IT = d0_total_filtered_IT.filter(items=['OBS_VALUE','TIME_PERIOD']).reset_index().drop(['index'], axis=1)
print(immigrants_per_year_IT)

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 5)

# Add x, y gridlines
ax.grid(visible = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)


plt.bar(immigrants_per_year_IT['TIME_PERIOD'], immigrants_per_year_IT['OBS_VALUE'])

# Show all years in x-axis
plt.xticks(immigrants_per_year_IT['TIME_PERIOD'])

plt.xlabel("years")
plt.ylabel("No. of imigrants")
plt.title("The number of imigrants in Italy over the years")
plt.show()
```
